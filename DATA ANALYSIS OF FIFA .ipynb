```python
#IMPORTING LIBRARIRES
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#to load the file
fifa=pd.read_csv("fifa data.csv")
```

```python
# to get the first five value fron the data
fifa.head()
```

```python
# to know the name of every column from the dataset
for col in fifa.columns:
    print(col)
```

```python
# to know how many columns and rows
# so rows are 18278 and columns are 104
fifa.shape
```

```python
#exploratory anlaysis:
# if want to count the no of players by their nationality
fifa['nationality'].value_counts()
```

```python
#top 10 country which has maximum player
fifa['nationality'].value_counts()[0:10]
```

```python
# top 5 country which has maximum player
fifa['nationality'].value_counts()[0:5]
```

```python
# if i wnat to plot a bar plot for top 5 countries then first sepearte key and values
fifa['nationality'].value_counts()[0:5].keys
```

```python
# for plotting bar plot
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="g")
plt.show()        
```

```python
# in this analysis if i want to extract columns from dataset
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
# if want to sort salary of player form high to less
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
# bar plot for players according to thier salary
#x-axis-players short_name and y-axis-wage_eur
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["grey","green","red","pink","orange"])
plt.show()
```
